```python
import pandas as pd
import plotly.express as px
from datetime import datetime
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
```

```python
#Month calculate

data = pd.read_csv("all_metro_employment.csv")
Figuredata = data.iloc[:,0:5]
Figuredata = Figuredata[["state","area","time","Mining, Logging and Construction"]]

Month = Figuredata['time'].values
Month = [datetime.strptime(m[0:3], '%b').month for m in Month]
Figuredata['Month'] = Month
Figuredata['month'] = Figuredata['Month']
Month = Figuredata['time'].values

Year = [y[7:11] for y in Month]
Figuredata['year'] = Year
Figuredata = Figuredata[Figuredata['year'] != "2021"]
Figuredata_temp = Figuredata[Figuredata['year'] != "2020"]

Figuredata
```

```python


Figuredata1 = Figuredata_temp.groupby(['month'])['Mining, Logging and Construction'].sum()
Figuredata1 = Figuredata1.reset_index()

fig1 = px.line(Figuredata1, x='month',y="Mining, Logging and Construction",labels=
                     {"Mining, Logging and Construction": "Employment Rate [in 1000s]",
                      "month":"Month",},
               markers=True)
              
fig1.update_xaxes(showline=True, linewidth=1, linecolor='black', mirror=True,type='category')
fig1.update_yaxes(showline=True, linewidth=1, linecolor='black', mirror=True)
fig1.for_each_yaxis(lambda a: a.update(title_text=a.title.text.replace("sum of", "")))
fig1.update_layout(bargap=0.2).show()
```

```python
# plot 2 revised
Figuredata_3 = Figuredata_temp.groupby(['state','month'])['Mining, Logging and Construction'].sum()
Figuredata_3 = Figuredata_3.reset_index()

fig = px.line(data_frame =Figuredata_3 , x= 'month', y= "Mining, Logging and Construction" ,markers = True,
              facet_col="state", facet_col_wrap=6,
              facet_col_spacing=0.05, facet_row_spacing = 0.02,
              labels={"month" : "Months",
              "Mining, Logging and Construction": "Employment Rate [in 1000s]"})

fig.update_layout(
    autosize=True, 
    width=2000,
    height=2500,
    margin=dict(
        l=50,
        r=50,
        b=10,
        t=50,
        pad=1
    ),)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1], font_size=14))
fig.update_xaxes(type='category',title_font = {"size": 12},showticklabels=True)
fig.update_yaxes(matches=None, showticklabels=True, visible=True, title_font = {"size": 13})
fig.show()
```

```python
from numpy import size
# plot 3 revised

Figuredata_3b = Figuredata_temp.loc[Figuredata_temp['area'].isin(["Washington-Arlington-Alexandria","New York-Newark-Jersey City","Chicago-Naperville-Elgin",
                                                             "New York-Jersey City-White Plains","Philadelphia-Camden-Wilmington",
                                                        "Houston-The Woodlands-Sugar Land","Los Angeles-Long Beach-Anaheim",
                                                             "Boston-Cambridge-Nashua","Dallas-Fort Worth-Arlington",
                                                        "Minneapolis-St. Paul-Bloomington"])]

Figuredata_3b = Figuredata_3b.groupby(['area','month'])['Mining, Logging and Construction'].sum()
Figuredata_3b = Figuredata_3b.reset_index()

fig = px.line(data_frame =Figuredata_3b , x= 'month', y= "Mining, Logging and Construction" ,markers = True,
              facet_col="area", facet_col_wrap=3,labels={"month" : "Months",
              "Mining, Logging and Construction": "Employment Rate [in 1000s]"},
              facet_col_spacing=0.1,
    facet_row_spacing=0.1)

fig.update_layout(
    autosize=True, 
    width=1000,
    height=1000,
    margin=dict(
        l=50,
        r=50,
        b=50,
        t=50,
        pad=1
    ),
)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_xaxes(type="category", title_font = {"size": 10},showticklabels=True)
fig.update_yaxes(matches=None, showticklabels=True, visible=True, title_font = {"size": 10})

fig.show()
                                                    
```
