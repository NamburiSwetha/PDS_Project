```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly
import plotly.graph_objects as go

dataFrame = pd.read_csv("all_metro_employment.csv")

dataFrame = dataFrame[["state", "area","time", "Mining, Logging and Construction"]]
```

```python
fig = px.histogram(dataFrame, x = 'time', y ="Mining, Logging and Construction", color ="state",
            labels={'time': ' ', 'Mining, Logging and Construction':'number of employments (grouped by 1000)'}, template="simple_white")
fig.update_xaxes(title_text="Time period")
fig.update_yaxes(title_text="Number of employments (grouped by 1000)")

fig.update_layout(
    autosize=False,
    width=1600,
    height=1200,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
)


fig.show()
```

```python
#covid data 2020 2021
covidData = dataFrame[(dataFrame['time'].str[-2:] == "20") | (dataFrame['time'].str[-2:] == "21")].reset_index(drop=True)
```

```python
StateTimeAgg = covidData.groupby(by =['state','time'])['Mining, Logging and Construction'].sum().to_frame(name = 'Mining, Logging and Construction').reset_index()

StateTimeAgg['time'] = pd.to_datetime(StateTimeAgg.time)

StateTimeAgg = StateTimeAgg.sort_values(by=['time', 'state'])

StateTimeAgg = StateTimeAgg.reset_index(drop = True)
```

```python
#pre covid data 2018 2019
preCovidData = dataFrame[(dataFrame['time'].str[-2:] == "18") | (dataFrame['time'].str[-2:] == "19")].reset_index(drop=True)

aggPreCovidData = preCovidData.groupby(by =['area'])['Mining, Logging and Construction'].sum().to_frame(name = 'Mining, Logging and Construction').reset_index()
aggPreCovidData = aggPreCovidData.reset_index(drop = True)

top10PreCovidAreas = aggPreCovidData.nlargest(10, ['Mining, Logging and Construction']).reset_index(drop = True)
```

```python
#april 2020 one month (covid) data
augCovidData = dataFrame[(dataFrame['time'].str[:] == "apr 1, 2020")].reset_index(drop=True)
```

```python
areaTimeAggCovid = augCovidData.groupby(by =['area','state'])['Mining, Logging and Construction'].sum().to_frame(name = 'Mining, Logging and Construction').reset_index()

top10CitiesCovid = areaTimeAggCovid.nlargest(10, ['Mining, Logging and Construction']).reset_index(drop = True)

top10CitiesCovid['timePeriod'] = "covid"
```

```python
#april 2019 one month (pre covid) data
augPreCovidData = dataFrame[(dataFrame['time'].str[:] == "apr 1, 2019")].reset_index(drop=True)
```

```python
areaTimeAggPreCovid = augPreCovidData.groupby(by =['area','state'])['Mining, Logging and Construction'].sum().to_frame(name = 'Mining, Logging and Construction').reset_index()

top10CitiesPreCovid = areaTimeAggPreCovid.nlargest(10, ['Mining, Logging and Construction']).reset_index(drop = True)

top10CitiesPreCovid['timePeriod'] = "preCovid"
```

```python
preCovidVsCovid = pd.concat([top10CitiesPreCovid, top10CitiesCovid], axis=0).reset_index(drop = True)
```
